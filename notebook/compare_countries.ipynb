Load Cleaned Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load cleaned data
benin = pd.read_csv('../data/benin-malanville-cleaned.csv')
togo = pd.read_csv('../data/togo-daaong_qc-cleaned.csv')
sierraleone = pd.read_csv('../data/sierraleone-bumbuna-cleaned.csv')

# Add country label
benin['Country'] = 'Benin'
togo['Country'] = 'Togo'
sierraleone['Country'] = 'Sierra Leone'

# Combine into one DataFrame
df = pd.concat([benin, togo, sierraleone], ignore_index=True)
```

Metric Comparison: Boxplots (Side-by-Side)

```python
metrics = ['GHI', 'DNI', 'DHI']
titles  = [
    'Global Horizontal Irradiance (GHI)',
    'Direct Normal Irradiance (DNI)',
    'Diffuse Horizontal Irradiance (DHI)'
]

plt.figure(figsize=(15, 10))

for i, metric in enumerate(metrics, start=1):
    # 2 rows Ã— 3 columns
    plt.subplot(2, 3, i)                                   
    sns.boxplot(
        data=df,
        # category on the x-axis
        x='Country',          
        y=metric, 
         # colour by country            
        hue='Country',       
        palette='Set2',
        legend=False          
    )
    plt.title(titles[i-1], fontsize=12, fontweight='bold')
    plt.ylabel('W/mÂ²')
    plt.xlabel('')

plt.tight_layout()
plt.show()
```
